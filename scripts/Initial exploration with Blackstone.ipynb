```python
import pandas as pd
import numpy as np
import os
import glob
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
path = r'/Users/shiyiming/Desktop/Green Investment/data/exploration'
bx_files = glob.glob(os.path.join(path, "*.csv"))
```

```python
bx = pd.concat((pd.read_csv(f,dtype={'CUSIP': str}) for f in bx_files), ignore_index=True)
```

```python
bx = bx.drop('Unnamed: 0', 1)
```

```python
bx['Period of report']=pd.to_datetime(bx['Period of report'], format='%Y-%m-%d')
```

```python
bx.dtypes
```

```python
bx = bx.sort_values(by=['Period of report'])
bx['value']=bx['VALUE (x$1000)'].div(1000)
```

```python
bx_noamend = bx.loc[bx['File type']=='Form 13F-HR']

fig, (ax1, ax2, ax3) = plt.subplots(3,1,figsize=(10,10), sharex=False)

bx_value = bx_noamend.pivot_table(index='Period of report', values='value', aggfunc='sum')
bx_value.plot(kind="line", figsize=(10,10), title="$m invested", ax=ax1)

bx_ncompany = bx_noamend.pivot_table(index='Period of report', values='NAME OF ISSUER', aggfunc='count')
bx_ncompany.plot(kind="line", figsize=(10,10), title="# Companies invested in", ax=ax2)

bx_nquarter = bx_noamend.pivot_table(index='CUSIP', columns = "Period of report", values="value", aggfunc='sum')
bx_nquarter["HeldFor"] = bx_nquarter.count(axis=1)

sns.histplot(bx_nquarter["HeldFor"], bins=30, ax = ax3)
ax3.set(title="# Total quarters a company has been held for (not continous)")

plt.tight_layout()
```
